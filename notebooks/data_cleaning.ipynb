```python
# Import pandas(for making data tables)
import pandas as pd
import numpy as np
from datetime import datetime

# Graphing modules
import matplotlib.pyplot as plt
from matplotlib import cm as cm
from matplotlib import mlab as ml
import seaborn as sns

# Makes plots in notebook
%matplotlib inline
```

```python
# Create a Dataframe (data from data.seattle.gov)
df = pd.read_csv('../data/Seattle_Real_Time_Fire_911_Calls.csv', low_memory=False)
```

```python
df.info()
```

```python
df.shape
```

```python
df = df[pd.notnull(df['Datetime'])]
df = df[pd.notnull(df['Longitude'])]
df = df[pd.notnull(df['Latitude'])]
df.info()
```

```python
df.head()
```

```python
# Type of 911 calls made with counts
df.groupby('Type').size()
```

```python
# Replace the spaces (missing date) in Latitude with NaN
df.Latitude = pd.to_numeric(df.Latitude, errors="coerce")
df.info()
```

```python
length = []
for dt in df.Datetime:
    if len(dt) not in length:
        length.append(len(dt))
length
```

```python
# Remove +0000
dt_list = []
for dt in df.Datetime:
    if 'T' in dt:
        dt_list.append(dt[:-5].strip())
    if 'AM' in dt or 'PM' in dt:
        dt_list.append(dt[:-6])
```

```python
df['dt_crop'] =  dt_list
df.head()
```

```python
date_list = []
for time in df.dt_crop:
    if 'T' in time:
        date_list.append(time[:10])
    else:
        month = time[:2]
        day = time[3:5]
        year = time[6:10]
        date = year + "-" + month + '-' + day
        date_list.append(date)  
```

```python
df['date'] = date_list
df.head()
```

```python
len(date_list)
```

```python
for time in df.dt_crop[0:50000]:
    if 'T' in time:
        print time [-8:]
```

```python
time_list = []
for time in df.dt_crop:
    if 'A' in time:
        time = time[-11:-3]
        time_list.append(time)
    elif 'P' in time:
        hr = time[-11:-9]
        hr = int(hr) + 12
        hr = str(hr)
        time = hr + time[-9:-3]
        time_list.append(time)
    elif 'T' in time:
        time = time[-8:]
        time_list.append(time)
```

```python
len(time_list)
```

```python
df['time'] = time_list
df.head()
```

```python
# Change hour 24 to hour 0 so can be cast as datetime
df['date2'] = pd.to_datetime(df.date)
df['time2'] = pd.to_timedelta(df.time)
# Adjust the date
df.Datetime = df.date2 + df.time2
df.info()
```

```python
# Drop duplicate rows
df_no_dup = df.drop_duplicates()
```
